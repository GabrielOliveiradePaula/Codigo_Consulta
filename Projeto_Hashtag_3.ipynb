```python
#1-Importando as bibliotecas - Pandas e Selenium 

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

#Selenium precisa de um outro código para usar, o from diz um livro da biblioteca Selenium
#Chrome: navegador utilizado
#Baixar crhomedriver 97 colocálo em uma pasta dentro da pasta que contém o python

#2-Abrindo o Chrome e pesquisar e pegar a cotação de dólar
navegador = webdriver.Chrome()
navegador.get("https://www.google.com/")

#Get(<site>): Método que realiza acesso à página

navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dólar")
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

#Depois de abrir o Chrome, localizar a barra de pesquisa, para isso, deve-se clickar com o botão direito do mouse; inspecionar; clickar na barra de pesquisa; clickar com botão direito no código; copy; copy full xpath
#Find_element: busca o método para selecionar a barra de pesquisa
#/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input: código Xpath
#.send_keys: digita o texto quando usa "" e aperta tecla quando usa Keys.

cotacao_dolar = navegador.find_element(By.XPATH, '/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div/div/div/div/div[3]/div[1]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(cotacao_dolar)

#'data-value': atributo mostrado pelo código HTML que recebe o valor cotação

#3-Abrindo o Chrome e pesquisar e pegar a cotação do euro

navegador.get("https://www.google.com/")
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação euro")
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
cotacao_euro = navegador.find_element(By.XPATH, '/html/body/div[7]/div/div[10]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div/div/div/div/div[3]/div[1]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(cotacao_euro)

#4-Pesquisar e pegar a cotação do ouro

navegador.get("https://www.melhorcambio.com/ouro-hoje")
cotacao_ouro = navegador.find_element(By.XPATH,'/html/body/div[5]/div[1]/div/div/input[2]').get_attribute("value")
cotacao_ouro = cotacao_ouro.replace(",",".")
print(cotacao_ouro)
navegador.quit()

#cotacao_ouro.replace(",","."): é preciso alterar separador de casas decimais para o do python (,para .)
#navegador.quit(): fecha o navegador

#5-Importando a planilha Excel

tabela = pd.read_excel(r'C:/Users/Computador/Desktop/Gabriel/Análises, Economia, Finanças, Empreendedorismo, Investimentos e Programação/Programação/Python/Hashtag/Produtos.xlsx')
display(tabela)

#6-Atualizando a planilha Excel

tabela.loc[tabela["Moeda"] == "Dólar", "Cotação"] = float(cotacao_dolar)
tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(cotacao_euro)
tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(cotacao_ouro)
display(tabela)

#tabela: modificação serpa no dataframe tabela
#.loc: permite selecionar apenas linhas que atendam ao requisito de ter "Dólar" na coluva Moeda
#Cotação: Encontradas as linhas que possuem “Dólar” na coluna “Moeda”, o Loc irá alterar a coluna “Cotação”

tabela["Preço Base Reais"] = tabela["Preço Base Original"] * tabela["Cotação"]
tabela["Preço Final"] = tabela["Preço Base Reais"] * tabela["Ajuste"]
display(tabela)

#As colunas que dependem das variáveis não foram alteradas junto, portanto devemos alterar

#8-Exportando os dados

tabela.to_excel("Produtos Novo.xlsx", index=False)

#Produtos Novo.xlsx: nome do arquivo criado
#index=False: Indica que não deverá ser exportada a coluna com o índice das linhas
```
